```python
# Importing Libraries
import json
import os
import random
from collections import Counter
```

```python
# User Model Class
class UserModel:
    def __init__(self, filepath="user_data.json"):
        self.filepath = filepath
        self.history = []
        self.load()

    def update(self, choice):
        self.history.append(choice)
        self.save()

    def get_preferences(self):
        return Counter(self.history)

    def most_common(self):
        prefs = self.get_preferences()
        return prefs.most_common(1)[0][0] if prefs else None

    def save(self):
        with open(self.filepath, "w") as f:
            json.dump(self.history, f)

    def load(self):
        if os.path.exists(self.filepath):
            with open(self.filepath, "r") as f:
                self.history = json.load(f)
```

```python
# Adaptive Engine
class AdaptiveEngine:
    def __init__(self, user_model):
        self.user_model = user_model
        self.content = {
            "movies": ["Toy Story", "Avatar", "Mystery Men"],
            "sports": ["Hockey", "Football", "Water Polo"],
            "tech": ["Cloud Computing", "Artificial Intellignece (AI)", "Machine Learning"]
        }

    def get_weighted_preferences(self):
    
        history = self.user_model.history
        weights = {}

        for i, choice in enumerate(history):
            # More recent choices get higher weight
            weight = (i + 1) / len(history)
            weights[choice] = weights.get(choice, 0) + weight

        return weights

    def recommend(self):
        history = self.user_model.history

        if not history:
            category = random.choice(list(self.content.keys()))
            return category, self.content[category]

        weighted_prefs = self.get_weighted_preferences()

        category = max(weighted_prefs, key=weighted_prefs.get)

        return category, self.content[category]
```

```python
# Interface (User Interaction Layer)
def run_system():
    user = UserModel()
    engine = AdaptiveEngine(user)

    print("=== Adaptive Recommendation System ===")

    while True:
        print("\nAvailable categories: movies, sports, tech")
        choice = input("Select a category (or type 'exit'): ").lower()

        if choice == "exit":
            print("Exiting system. Goodbye!")
            break

        if choice not in ["movies", "sports", "tech"]:
            print("Invalid choice. Try again.")
            continue

        # Update user model
        user.update(choice)

        # Generate adaptive recommendation
        category, recommendations = engine.recommend()

        print(f"\n[Adaptive System]")
        print(f"Based on your behavior, you might like: {category}")
        print("Recommendations:", ", ".join(recommendations))
```

```python
# Running the System
run_system()
```
